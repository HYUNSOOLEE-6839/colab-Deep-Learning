# 선형 회귀 - 보스턴 집값 예측

```
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```
# 실행할 때 마다 같은 결과를 출력하기 위한 seed 값 설정
seed = 2021
np.random.seed(seed)
tf.random.set_seed(seed)
```

```
from sklearn.datasets import load_boston
boston = load_boston()
```

```
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, random_state=seed
)
X_train.shape, y_train.shape
```

```
model = Sequential()
model.add(Dense(30, input_dim=1, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
model.summary()
```

```
model.compile(loss='mean_squared_error', 
              optimizer='adam')
# 이전에 크로스엔트로피를 사용한 것과 달리 에러를 제곱한 것의 평균을 로스함수로 주게됨.
```

```
model.fit(X_train, y_train, epochs=500, batch_size=10)
```

### 평가

##### 밑에서 쓰일 flatten 함수는 list를 array로 만들어줌.

```
# 예측 값과 실제 값의 비교
y_prediction = model.predict(X_test).flatten()
for i in range(10):
  label = y_test[i]
  prediction = y_prediction[i]
  print(f"실제가격: {label:.3f}, 예상가격: {prediction:.3f}")
```
